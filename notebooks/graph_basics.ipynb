```python
"""lets start here u see  me learning this networkx the graph analyzer python package """
import networkx as nx 
import matplotlib.pyplot as plt
G = nx.Graph();
G.add_node("A");
G.add_nodes_from(["B","C","D"])
G.add_edge("A","B")
G.add_edges_from([("A","C"),("C","D")])
nx.draw(G,with_labels = True,node_color = ["red","blue","red","blue"],node_size = 600)
plt.show()
print("Nodes:", G.nodes())
print("Edges:", G.edges())
print("Neighbors of A:", list(G.neighbors("A")))
```

```python
import networkx as nx 
import matplotlib.pyplot as plt
Gr = nx.Graph()
Gr.add_nodes_from(["A","B","C","D"])
Gr.add_weighted_edges_from([("A","B",10),("A","C",9),("B","D",6),("C","D",11)])
path = nx.shortest_path(Gr, source = "A",target = "D",weight = "weight")
print(path)
print(list(Gr.degree()))


```

```python
import networkx as nx
import matplotlib.pyplot as plt 
import random
"""here's the deal i dont have any real data  so i need to create some lets go """
G = nx.erdos_renyi_graph(20, 0.12)
"""just an model to create iur fake data 20->nodes 0.12 -> probality """
for node in G.nodes():
    G.nodes[node]["vul"]= round(random.random(),2)
    G.nodes[node]["critical"]= random.choice([0,0,0,1])
pos = nx.spring_layout(G,seed = 42)
colors = []
for n in G.nodes():
    if G.nodes[n]['critical'] == 1:
        colors.append('red')
    else:
        colors.append('green')
nx.draw(G,pos,node_color = colors,with_labels  = True)
plt.show()
```
